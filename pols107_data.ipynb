### Cleaning Data

```python
#importing in the CSV
df = pd.read_csv("POLS 172_ Dset2.csv")
```

```python
#renaming the columns
df = df.rename(columns={"Retweet, Post, Quotetweet?": "typepost"})
df = df.rename(columns={"Party ": "Party"})
df = df.rename(columns={"Political Action Tweet Clasification ":"political_action"})
df = df.rename(columns={"Israel/Hamas Tweet Classification ":"israel_hammas"})
df = df.rename(columns={"Ukraine/Russia Tweet Classification":"ukraine_russia"})
df = df.rename(columns={"Economic Tweet Classification":"economy"})
df = df.rename(columns={"Trump Trials Tweet Classication":"trump_trials"})
df = df.rename(columns={"Immigration Tweet Classification":"immigration"})
df = df.rename(columns={"Climate Change Tweet Classification":"climate_change"})
df = df.rename(columns={"Local Issues Tweet Classification":"Local_issues"})
df = df.rename(columns={"Unnamed: 0":"Name"})
df = df.rename(columns={"Abortion Rights":"abortion_rights"})
df = df.rename(columns={"Gun Violence":"gun_issues"})
df = df.rename(columns={"Partisan Rivalry":"partisan_rivalry"})
df = df.rename(columns={"Up for reelection?":"reelection?"})
```

```python
df.columns
```

```python
#here im just subssetting the columns that I want to try to work with"
df = df[["Name", 'State', 'Party', 'Twitter Handle', 'Date',
       'political_action', 'israel_hammas', 'ukraine_russia', 'economy',
       'trump_trials', 'immigration', 'climate_change', 'Local_issues',
       'Good Will', 'Abortion Rights', 'Gun Violence', 'Partisan Rivalry',
       'Other', 'Up for reelection?','Tweet Link',
       'typepost']]
```

```python
#trying to fix the date column with a function
def date_function(datestring):
    """ This is the function that makes the date column into a workable date time string """
    try:
        return pd.to_datetime(datestring, format = "%Y%m%d")
    except ValueError:
        return print("Not working")
```

```python
#applying the funcion to the column
df["Date"] = df["Date"].apply(date_function)
#df["Date"] = df["Date"].fillna(0)
```

```python
#I am dropping the empty data cells in the local_issues column so I can change the dtype and work with
#df["Local_issues"] = df["Local_issues"].replace(["", ], 0)
#df["Local_issues"] = df["Local_issues"].fillna(0)
df['Local_issues'] = df['Local_issues'].astype(int)
```

```python
df.dtypes
```

```python
#Cleaning the partisan rivalry column
df["Partisan Rivalry"] = df['Partisan Rivalry'].fillna(0, inplace=True)
df["Partisan Rivalry"] = df["Partisan Rivalry"].replace(["-"], 0)
df['Partisan Rivalry'] = df['Partisan Rivalry'].astype(int)
```

```python
#creating a column that makes 'party' into integers to analyze 
def party_num(string):
    if string == 'Republican ':
        return 1 
    else:
        string == "Democrat"
        return 0
        
df["party_num"] = df["Party"].apply(party_num)
```
